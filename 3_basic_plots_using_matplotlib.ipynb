# 3 basic plots using matplotlib
               importing matplotlib package

working with **matplotlib** to create graph

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
fig = plt.figure()
axe = plt.axes()
```

creating a wave in graph

```python
fig = plt.figure()
axe = plt.axes()
x = np.linspace(0, 10, 1000)
axe.plot(x, np.sin(x))
```

# Labeling plots:
    Titles and axis labels are the simplest method to labeling the graph

```python
plt.plot(x,np.sin(x)) 
plt.title('A sine curve') 
plt.xlabel('x') 
plt.ylabel('sin(x) ')
```

**Adjusting the plot axes limits**

```python
plt.plot(x,np.sin(x)) 
plt.xlim(-1, 10) 
plt.ylim(-3.5, 3.5)
```

**Multiple lines in a single axes**

```python
plt.plot(x,np.sin(x), '-g', label='sin(x) ') 
plt.plot(x,np.cos(x), ':b', label='cos(x) ') 
plt.legend()
```

Creating **scatter plots** using matplotlib

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='green')
```

crating **randam plots** in scatter plots using matplotlib

```python
ang = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^','d']:
    plt.plot(ang.rand(5), ang.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```
