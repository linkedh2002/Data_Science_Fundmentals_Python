# **Scatter markers**

Importing the Required Modules

```python
from bokeh.plotting import figure, output_file, show, output_notebook
```

Creating a figure,
Put output_notebook() to Get the notebook,
Add a circle render with Size,Color and Alpha

```python
p = figure(plot_width = 400, plot_height = 400)
output_notebook() 
p.circle([1, 2, 3, 4, 5], [4, 7, 1, 6, 3], 
         size = 10, color = "red", alpha = 0.5)
show(p)
```

Similarly, use the square() method of Figure to scatter square markers on a plot

```python
p = figure(plot_width = 400, plot_height = 400)
output_notebook() 
p.square([1, 2, 3, 4, 5], [4, 7, 1, 6, 3], 
         size = 10, color = "green", alpha = 1)
show(p)
```

To create a single line Graph, line() method is used.

```python
p = figure(plot_width = 400, plot_height = 400)
output_notebook()
p.line([1, 2, 3, 4, 5], [3, 1, 2, 6, 5], 
        line_width =2, color = "red")
show(p)
```

# **Step lines**

To get discrete steps between data points, we use step() glyph method

```python
output_notebook()
p = figure(width=400, height=400)
p.step([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2, mode="center")
show(p)
```

# **Multiple lines**

If you want to draw multiple lines, use the multi_line() glyph method

```python
p = figure(width=400, height=400)

p.multi_line([[1, 3, 2], [3, 4, 6, 6]], [[2, 1, 4], [4, 7, 8, 5]],
             color=["firebrick", "navy"], alpha=[0.8, 0.3], line_width=4)

show(p)
```

# **Bar Graph**

To draw vertical bars by specifying a center x-coordinate, width, and top and bottom endpoints, use the vbar() glyph Method

```python
p = figure(width=400, height=400)
p.vbar(x=[1, 2, 3], width=0.5, bottom=0,
       top=[1.2, 2.5, 3.7], color="firebrick")

show(p)
```

To draw horizontal bars by specifying a center y-coordinate, height, and left and right endpoints, use the hbar() glyph Method

```python
p = figure(width=400, height=400)
p.hbar(y=[1, 2, 3], height=0.5, left=0,
       right=[1.2, 2.5, 3.7], color="navy")

show(p)
```
