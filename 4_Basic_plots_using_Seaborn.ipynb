# Importing the required modules(Matplotlib,Pandas,Numpy)With convenient Names

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
```

### Obtaining Data to Creat Graphs

```python
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

# Using Seaborn to creat datasets

```python
df = sns.load_dataset('titanic')
df.head()
```

# Expressing The dataset as Bar Graph

```python
sns.countplot(x='sex',data=df)
```

# Using count plot and customizing the Graph

```python
sns.countplot(x='sex', hue = 'survived', data = df,
palette = 'Set1')
```

# Using KDE plot to plot continous data

```python
sns.kdeplot(x = 'age' , data = df , color = 'cyan')
```

# Plotting continous data but using displot

```python
sns.displot(x = 'age',kde=True,bins = 5 , data =df)
```

# Loading the dataset Iris

```python
df = sns.load_dataset('iris')
df.head()
```

# Expressing the graph with scatterplot

```python
sns.scatterplot(x='sepal_length', y ='petal_length' ,
data = df , hue = 'species')
```

# Expressing the graph using jointplot

```python
sns.jointplot(x='sepal_length' , y ='petal_length',
data = df, kind='reg' )
```

# We can plot multiple scatter plots using pairplot

```python
sns.pairplot(df)
```

# **Cat** plots can be used visualize the datas et as parameter with bins
